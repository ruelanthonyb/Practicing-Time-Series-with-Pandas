```python
import pandas as pd
url="https://docs.google.com/spreadsheets/d/e/2PACX-1vT_jChgNsQbHbg4TGepzIqk8XC9DTIKmyyxb1upo5cfZCgbfIUQc2ZC0YMzuU5uApP140Ob49KBjdqh/pub?gid=1198589591&single=true&output=csv"
df = pd.read_csv(url)
df.info()
df.head(3)
```

### 1. Convert the date column to datetime dtype.

```python
pd.to_datetime(df['date'])
```

```python
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')
df.info()
df.head(3)
```

### 2. Set the date column as the index.

```python
df = df.set_index('date')
df
```

```python
df.index
```

### 3. Filter out only the required data:
* Keep only data from the year 2000 or later.
* Keep only the following features:
    * precipitation
    * mean_temp
    * min_temp
    * max_temp
    * snow_depth

```python
filtered_df = df.loc['2000':, ['precipitation','mean_temp', 'min_temp', 'max_temp', 'snow_depth']]
filtered_df
```

### 4. Answer the following questions.

a) "The lowest temperature of {max temp} degrees occurred on {date as "MonthName 2-digit-day, 4-digit year.} which was a {day of the week}."

```python
min_temp_date = filtered_df['min_temp'].idxmax()
min_temp = filtered_df['min_temp'].max()
fmt = "%B %d, %Y"

print(f"The lowest temperature of {min_temp} degrees occurred on {min_temp_date.strftime(fmt)}, \
which was a {min_temp_date.day_name()}.")
```

b) "The highest daily precipitation of {max preciptation} inches occurred on {date as "MonthName 2-digit-day, 4-digit year.} which was a {day of the week}."

```python
max_precip_date = filtered_df['precipitation'].idxmax()
max_precip = filtered_df['precipitation'].max()
fmt = "%B %d, %Y"

print(f"The highest daily precipitation of {max_precip} inches occurred on {max_precip_date.strftime(fmt)}, \
which was a {max_precip_date.day_name()}.")
```
